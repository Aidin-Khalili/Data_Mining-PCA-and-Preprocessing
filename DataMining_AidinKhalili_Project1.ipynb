```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler , MinMaxScaler
from sklearn.decomposition import PCA


def removeDublicateSamples(theDataset):
    theDataset.drop_duplicates(inplace=True)

def removeNonNumericFeatures(theDataset):
    return theDataset.select_dtypes(include=['int64', 'float64'])

def identifyMissingValues(theDataset):
    for i in range(len(theDataset.columns)):
        theDataset.iloc[:,i].fillna(theDataset.iloc[:,i].mean(), inplace=True)

def dropClass(theDataset, theClassifiersLabel):
    theDataset.drop(axis=1,labels=[theClassifiersLabel], inplace=True)
        
 
```

```python
url_BrCaWi = 'https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data'
BrCaWi = pd.read_csv(url_BrCaWi, names=['Sample code number', 'Clump Thickness', 'Uniformity of Cell Size', 'Uniformity of Cell Shape', 'Marginal Adhesion', 'Single Epithelial Cell Size', 'Bare Nuclei', 'Bland Chromatin', 'Normal Nucleoli', 'Mitoses', 'Class'])

# Preprocessing

removeDublicateSamples(BrCaWi)

dropClass(BrCaWi, 'Class')

BrCaWi = removeNonNumericFeatures(BrCaWi)

identifyMissingValues(BrCaWi)

scaler_BrCaWi = StandardScaler()
scaler_BrCaWi.fit(BrCaWi)
scaled_BrCaWi = scaler_BrCaWi.transform(BrCaWi)

# PCA
pca_BrCaWi = PCA(n_components=2)
pca_BrCaWi.fit(scaled_BrCaWi)
x_pca_BrCaWi = pca_BrCaWi.transform(scaled_BrCaWi)

plt.figure(figsize=(13,4))
sns.heatmap(pca_BrCaWi.components_)
```

```python
from scipy.io import  arff
from urllib.request import urlopen
from io import StringIO

url_DiReDe = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00329/messidor_features.arff'
response = urlopen(url_DiReDe)
data, meta = arff.loadarff(StringIO(response.read().decode('utf-8')))
DiReDe =pd.DataFrame(data)

# Preprocessing

removeDublicateSamples(DiReDe)

dropClass(DiReDe, 'Class')

DiReDe = removeNonNumericFeatures(DiReDe)

identifyMissingValues(DiReDe)

scaler_DiReDe = StandardScaler()
scaler_DiReDe.fit(DiReDe)
scaled_DiReDe = scaler_DiReDe.transform(DiReDe)

# PCA
pca_DiReDe = PCA(n_components=2)
pca_DiReDe.fit(scaled_DiReDe)
x_pca_DiReDe = pca_DiReDe.transform(scaled_DiReDe)

plt.figure(figsize=(13,4))
sns.heatmap(pca_DiReDe.components_)
```
